```python
import sqlite3
import pandas as pd
# Create your connection.
import patents_lib as kPatents
MaDB = "patents.db"
```

# Loading DB as Pandas dataframe

```python
cnx = sqlite3.connect(MaDB)
df = pd.read_sql_query("SELECT * FROM patents", cnx)
df
```

```python
df["BW"] = df.p_backward.apply(lambda x: str(x).split(", "))
df["FW"] = df.p_forward.apply(lambda x: str(x).split(", "))
```

```python
AllSaved = df.p_id.to_list()
df
```

```python
df["Lower"] = df.p_abstract.apply(lambda x: x.lower())
DF = df[df['Lower'].str.contains("sector")]
DF.head()
```

```python
BWlist = list(filter(None,(", ".join(DF.p_backward.tolist())).split(", ")))
FWlist = list(filter(None,(", ".join(DF.p_forward.tolist())).split(", ")))

BackCount = pd.Series(BWlist).value_counts().reset_index()
BackCount.columns = ["p_id","countBW"]
BC = BackCount.merge(df,how="outer")
ToGetRetro = BC[(BC.countBW > 0) & (BC.p_link.isna())].head(20).p_id.to_list()
ToGetRetro
```

```python
for Patent in ToGetRetro:
    kPatents.get_soup(Patent,MaDB)
```

# Same thing for forward looking patents

```python
FWlist = list(filter(None,(", ".join(DF.p_forward.tolist())).split(", ")))

BackCount = pd.Series(FWlist).value_counts().reset_index()
BackCount.columns = ["p_id","countFW"]
BC = BackCount.merge(df,how="outer")
ToGetForward = BC[(BC.countFW > 0) & (BC.p_link.isna())].head(20).p_id.to_list()
ToGetForward
```

```python
for Patent in ToGetForward:
    kPatents.get_soup(Patent,MaDB)
```
