```python
import sqlite3
import pandas as pd
# Create your connection.
import patents_lib as kPatents
MaDB = "patents.db"
```

# Loading DB as Pandas dataframe

```python
cnx = sqlite3.connect(MaDB)
df = pd.read_sql_query("SELECT * FROM patents", cnx)
df["p_title"] = df.p_title.apply(lambda x: "-".join(x.split("-")[1:-1]).strip() if "Google" in x else x)
blacklist = "meat cutter surgical laxity comestic aesthetic dental toothbrush tooth cellulite tighten neck laser flossing flosser oral"
blacklist = blacklist.split(" ")
dfLower = df.p_title.str.lower()
df = df[~dfLower.str.contains('|'.join(blacklist))]
df
```

```python
df["Lower"] = df.p_abstract.apply(lambda x: x.lower())
DF = df[df['Lower'].str.contains("low power")]
DF.head()
```

```python
BWlist = list(filter(None,(", ".join(DF.p_backward.tolist())).split(", ")))
FWlist = list(filter(None,(", ".join(DF.p_forward.tolist())).split(", ")))

BackCount = pd.Series(BWlist).value_counts().reset_index()
BackCount.columns = ["p_id","countBW"]
BC = BackCount.merge(df,how="outer")
ToGetRetro = BC[(BC.countBW > 0) & (BC.p_link.isna())].head(20).p_id.to_list()
ToGetRetro
```

```python
for Patent in ToGetRetro:
    kPatents.get_soup(Patent,MaDB)
```

# Same thing for forward looking patents

```python
FWlist = list(filter(None,(", ".join(DF.p_forward.tolist())).split(", ")))

BackCount = pd.Series(FWlist).value_counts().reset_index()
BackCount.columns = ["p_id","countFW"]
BC = BackCount.merge(df,how="outer")
ToGetForward = BC[(BC.countFW > 0) & (BC.p_link.isna())].head(20).p_id.to_list()
ToGetForward
```

```python
for Patent in ToGetForward:
    kPatents.get_soup(Patent,MaDB)
```

# Saving DF to a CSV that is compressed

```python
df["p_title"] = df.p_title.apply(lambda x: "-".join(x.split("-")[1:-1]).strip() if "Google" in x else x)
dfLower = df.p_title.str.lower()
df = df[~dfLower.str.contains('|'.join(blacklist))]

df.to_csv("AllPatentsStructure.csv.gz", 
           index=False, 
           compression="gzip")
```

# Making pages



```python
topics = ["electromechanical"]
```

```python
df["BW"] = df.p_backward.apply(lambda x: str(x).split(", "))
df["FW"] = df.p_forward.apply(lambda x: str(x).split(", "))
AllSaved = df.p_id.to_list()
```

```python
for k in range(len(df)):
    sID = df.iloc[k]["p_id"]
    sTitle = df.iloc[k]["p_title"]
    sDate = df.iloc[k]["p_date"]
    sInventor = df.iloc[k]["p_inventor"]
    sBenef = df.iloc[k]["p_beneficiary"]
    sAbstract = df.iloc[k]["p_abstract"]
    sBW = df.iloc[k]["BW"]
    sFW = df.iloc[k]["FW"]
    
    MD = "# "+sID+" - "+sTitle.replace("- Google Patents","").replace("\n","").strip()+"\n\n"
    MD += "## Details\n\n"
    MD += "* Date: "+sDate.strip()+"\n"
    MD += "* Inventor: "+sInventor.strip()+"\n"    
    MD += "* Beneficiary: "+sBenef.strip()+"\n"  
    MD += "\n## Other patents\n"   
    if len(sBW):
        MD += "\n### Backwards\n * "
        MD += "\n * ".join(sBW)
    if len(sFW):
        MD += "\n### Forward\n * "
        MD += "\n * ".join(sFW)
    MD += "\n## Abstract\n\n"   
    MD += sAbstract.strip()
    
    for Referenced in AllSaved:
        MD = MD.replace(Referenced, "["+Referenced+"]("+Referenced+".md)")
        
    MD = "Source: ["+sID+"](https://patents.google.com/patent/"+sID+")" + "\n\n"+MD
    
    f= open("md/"+sID+".md","w+")
    f.write(MD)
    f.close()
```
